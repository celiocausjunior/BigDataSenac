```python
df = spark.read.csv("/content/sample_data/train.csv", header = True, inferSchema = True)
```

```python
df.show()
```

```python
df.count()
```

```python
df.select("User_ID", "Gender", "Age", "Occupation", "Purchase").show (5)
```

```python
df.select("Gender","Purchase").describe().show()
```

```python
df.select("Stay_In_Current_City_Years").distinct().show()

df.select("Age").distinct().show()

df.select("Occupation").distinct().show()

df.select("City_Category").distinct().show()

df.select("Marital_Status").distinct().show()

df.select("Product_Category_1").distinct().show()

df.select("Product_Category_2").distinct().show()

df.select("Product_Category_3").distinct().show()
```

```python
from pyspark.sql import functions as F

df.groupBy("City_Category").agg(F.sum("Purchase")).show()
```
